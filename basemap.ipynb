```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

# Drawing the first map

```python
# wthout passing any option, the map has the plate Caree projection 
# centered at longtitude and latitude = 0
map = Basemap()     
# after setting the map, we can draw what we want
map.drawcoastlines()
# open a window to explore the result 
plt.show()
```

```python
map = Basemap(projection="ortho", lat_0 = 0, lon_0 = 0)
# fill the globe with a blue color
map.drawmapboundary(fill_color='aqua')
# fill the continents with the land color
map.fillcontinents(color="coral", lake_color="aqua")

map.drawcoastlines()
plt.show()
```

# Managing Projections
Projection and its features are all assigned when the object Basemap is created. 

## Projection

```python
# default value: cyl - cylindrical Equidistant projection
map = Basemap(projection = 'cyl')
map.drawmapboundary(fill_color="aqua")
map.fillcontinents(color="coral", lake_color="aqua")
map.drawcoastlines()
plt.show()
```

```python
# pojection centered at longtitude = 10; latitude = 50; over Europe
map = Basemap(projection = 'aeqd', lon_0 = 10, lat_0 = 50)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()
plt.show()
```

## Using epsg to set the projection
epsg code is standard to name projections using a numerical code
can see epsg exists at 
`<python_packages_path>/mpl_toolkits/basemap/data/epsg`

```python
map = Basemap(llcrnrlon=3.75, llcrnrlat=39.75, urcrnrlon=4.35, urcrnrlat=40.15, 
              resolution='h', epsg = 5520)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()
plt.show()
```

## Passing the bounding box

```python
# lower-left and upper-right corners are past as parameters, 
# in longitude-latitude units, not in map units
map = Basemap(llcrnrlon=-10.5, llcrnrlat=35, urcrnrlon=4., urcrnrlat=44., 
              resolution='i', projection='tmerc'0, lat_0=39.5, lon_0=-3.25)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()

plt.show()
```

## Passing the bounding box in the map coordinates
- some projection-look like a satellite image-accept setting the extension using the map coordinates.
- the projection parameters must be set-center point-and then the zone to show can be only a part of the globe. 
- Only ortho, geos and nsper projections can use this method to set the map extension

```python
map = Basemap(resolution='l', 
              satellite_height = 3000000.,
              projection = 'nsper',
              lat_0 = 30., lon_0 = -27.,
              llcrnrx=500000., llcrnry=500000., urcrnrx=2700000., urcrnry=2700000.
              )
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()
plt.show()
```

## Passing the center, width and height
- the center of projection, the width and height of the projection are passed as parameters
- center is easy, just pass it in lon-lat
- units are the projection units in meters. 
- the point (0,0) => the lower left corner
- the point (width, height) => upper right
- origin of the positions is not the one defined by the projection as in GDAL

```python
map = Basemap(projection = 'aeqd', 
              lon_0 = 0,
              lat_0 = 90,
              width = 10000000,
              height = 10000000)
map.drawmapboundary(fill_color="aqua")
map.fillcontinents(color='coral', lake_color="aqua")
map.drawcoastlines()
for i in range(0, 10000000, 10000000):
    map.plot(i, i, marker='o', color='k')
plt.show()
```

# Basic Functions

## Drawing a point in a map

```python
map = Basemap(projection='ortho',
              lat_0=0, lon_0=0)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines()

x, y = map(0,0)
map.plot(x, y, marker='D', color='m')
plt.show()
```

- use the basemap instance to calculate the position of the point in the map coordinates when you hae the longitude and latitude of the point 
- the plot method needs the x and y position in the map coordinates, the marker and the color
- if you have more than one point, you may preder the scatter method.
